```python
import tensorflow as tf
tf.get_logger().setLevel('ERROR')
tf.autograph.set_verbosity(2)
from matplotlib import pyplot  
from utils import *
```

```python
# Load eval data
ds = load_eval_tfrecord('tfrecord_example/',512,1024,512,1024, False)
# Load pretrained model
model = tf.keras.models.load_model('ckpts/ZeroScatter/')
```

```python
capture, capture_proc = next(ds)
output = model(capture, training = False)
pyplot.figure(figsize=(15,20))
pyplot.subplot(2,1,1)
pyplot.imshow(tf.squeeze(capture))
pyplot.title('Capture')
pyplot.axis('off')
pyplot.subplot(2,1,2)
pyplot.imshow(tf.squeeze(tf.clip_by_value(output,0,1)))
pyplot.title('ZeroScatter')
pyplot.axis('off')
```
